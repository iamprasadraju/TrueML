```python
# loading datasets
import pandas as pd
import numpy as np

df = pd.read_csv("../../datasets/housing.csv")

features_names = [
    "longitude",
    "latitude",
    "housing_median_age",
    "total_rooms",
    "total_bedrooms",
    "population",
    "households",
    "median_income",
]

X = df[features_names]
y = df["median_house_value"]

X["total_bedrooms"] = X["total_bedrooms"].fillna(
    X["total_bedrooms"].mean()
)
```

```python
print(X.isnull().sum())
```

```python
print(np.isnan(X).any())
print(np.isnan(y).any())
```

```python
# Standardize each feature
X = (X - X.mean()) / X.std()
```

```python
# Training using TrueML LinReg
from trueml.linearmodel import LinearRegression
from trueml.losses import MSEloss
from trueml.visualize import LivePlot
import numpy as np

plot = LivePlot("Training Loss")
model = LinearRegression(n_features = X.shape[1], lr=0.001)
loss_fn = MSEloss()

for epoch in range(1000):
    y_pred = model.forward(X)

    t_loss = loss_fn(y, y_pred)
    d_loss = loss_fn.grad(y, y_pred)

    dw, db = model.grad(X, d_loss)

    model.backward(dw, db)

    if epoch % 100 == 0:
        plot(t_loss)
    
```
