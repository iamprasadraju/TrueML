```python
import numpy as np
import trueml.plots as plt

corr = np.array([
    [1.00, 0.82, -0.35],
    [0.82, 1.00, -0.12],
    [-0.35, -0.12, 1.00],
])

plt.heatmap(
    corr,
    title="Correlation Matrix",
    xticklabels=["Income", "Rooms", "Age"],
    yticklabels=["Income", "Rooms", "Age"], figsize=(5, 4)
)
```

```python
cov = np.array([
    [2.4, 1.2, 0.3],
    [1.2, 4.1, 0.8],
    [0.3, 0.8, 1.6],
])

plt.heatmap(
    cov,
    title="Covariance Matrix",
    xticklabels=["A", "B", "C"],
    yticklabels=["A", "B", "C"],
)
```

```python
cm = np.array([
    [50, 3],
    [4, 43],
])

plt.heatmap(
    cm,
    title="Confusion Matrix",
    xticklabels=["Pred 0", "Pred 1"],
    yticklabels=["True 0", "True 1"],
)
```

```python
dist = np.array([
    [0.0, 2.1, 3.5],
    [2.1, 0.0, 1.7],
    [3.5, 1.7, 0.0],
])

plt.heatmap(
    dist,
    title="Distance Matrix",
    xticklabels=["A", "B", "C"],
    yticklabels=["A", "B", "C"],
)
```
