```python
# loading datasets
import pandas as pd
import numpy as np

df = pd.read_csv("../datasets/housing.csv")

features_names = [
    "longitude",
    "latitude",
    "housing_median_age",
    "total_rooms",
    "total_bedrooms",
    "population",
    "households",
    "median_income",
]

X = df[features_names]
y = df["median_house_value"]

X["total_bedrooms"] = X["total_bedrooms"].fillna(
    X["total_bedrooms"].mean()
)
```

```python
print(X.isnull().sum())
```

```python
print(np.isnan(X).any())
print(np.isnan(y).any())
```

```python
# Training using TrueML LinReg

from trueml.linearmodel import LinearRegression
from trueml.losses import MeanSquaredError
from trueml.visualize import viz,LivePlot


model = LinearRegression(n_features = X.shape[1], lr=0.001)
loss_fn = MeanSquaredError()

plot = LivePlot()

for epoch in range(10000):
    y_pred = model.forward(X)
    t_loss = loss_fn(y, y_pred)
    d_loss = loss_fn.grad(y, y_pred)
    dw, db = model.grad(X, d_loss)
    model.backward(dw, db)

    
    if epoch % 100 == 0:
        plot.update(t_loss)

    
```
